# Predicción de Burnout en Estudiantes

## Descripción del dataset

El dataset *Student Mental Health and Burnout* contiene 150.000 registros de estudiantes universitarios con información sobre su estilo de vida, rendimiento académico y salud mental. Incluye variables como horas de sueño, nivel de estrés, presión académica, actividad física, tiempo de pantalla, entre otras.

## Problema a resolver

El objetivo es predecir si un estudiante va a sufrir burnout (alto o bajo) en base a sus características de vida y estudio. Esto permitiría a las instituciones educativas identificar estudiantes en riesgo e intervenir a tiempo antes de que el problema se agrave.

## Contexto Big Data

Este problema se enmarca en Big Data por las tres dimensiones clásicas:

- **Volumen**: el dataset contiene 150.000 registros, representando una escala de datos que requiere herramientas de procesamiento distribuido como Apache Spark.
- **Variedad**: combina datos numéricos (edad, horas de sueño, CGPA) y categóricos (género, carrera, calidad de sueño, nivel de estrés), lo que exige un preprocesamiento heterogéneo.
- **Velocidad**: en un sistema real, los datos llegarían de forma continua a medida que los estudiantes completan encuestas, requiriendo procesamiento en tiempo real o near real-time.

## Análisis Exploratorio de Datos (EDA)

Antes de entrenar cualquier modelo es fundamental entender el dataset. En esta sección analizamos:
- Dimensiones y tipos de datos
- Valores nulos
- Distribución del target `burnout_level`
- Estadísticas descriptivas de las variables numéricas

```python
import pandas as pd
import matplotlib.pyplot as plt

CSV_PATH = "/Volumes/workspace/default/trabajo-practico/student_mental_health_burnout.csv"
df = pd.read_csv(CSV_PATH)

# Dimensiones
print(f"Filas: {df.shape[0]} | Columnas: {df.shape[1]}")
print()

# Tipos de datos
print("Tipos de datos:")
print(df.dtypes)
print()

# Valores nulos
print("Valores nulos por columna:")
print(df.isnull().sum())
print()

# Distribucion del target
print("Distribución del target (burnout_level):")
print(df['burnout_level'].value_counts())
print()
df['burnout_level'].value_counts().plot(kind='bar', title='Distribución de Burnout Level')
plt.ylabel('Cantidad de estudiantes')
plt.xticks(rotation=0)
plt.show()

# Estadisticas basicas
print("Estadísticas descriptivas:")
df.describe()
```
